# ***SURPRISE LAB***



*Surprise* adalah *library* Python untuk membangun dan menganalisis sistem rekomendasi yang berkaitan dengan data *rating* eksplisit. *Surprise* dirancang dengan tujuan memberikan pengguna kendali yang sempurna atas eksperimen yang dilakukan, mengatasi masalah penanganan *dataset*, serta menyediakan berbagai algoritma prediksi siap pakai dan evaluasi untuk mengukur kinerja algoritma.

```python
pip install scikit-surprise
```

```python
# Import library Surprise
from surprise import Dataset, NormalPredictor, Reader
from surprise.model_selection import cross_validate, KFold
from surprise import SVD, accuracy
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
# Reader diperlukan dengan parameter rating_scale
reader = Reader(rating_scale=(1, 5))

# Kolom harus sesuai dengan user, item, dan rating
data = Dataset.load_from_df(df[["AuthorId", "RecipeId", "Rating"]], reader)
```

# **ALGORITMA SVD**



Algoritma *Singular Value Decomposition* (SVD) adalah teknik penguraian matriks yang digunakan dalam sistem rekomendasi untuk mengurai preferensi pengguna terhadap *item* ke dalam matriks pengguna, matriks *item*, dan matriks faktor tersembunyi. SVD memungkinkan sistem rekomendasi untuk membuat rekomendasi *item* berdasarkan faktor-faktor tersembunyi yang terkait dengan preferensi pengguna.

```python
# Cross-validation iterator
iterator = KFold(n_splits=3)
```

```python
# Menggunakan algoritma SVD
algo = SVD()
```

***Split Data***

```python
for trainset, testset in iterator.split(data):

    # train test
    algo.fit(trainset)
    predictions = algo.test(testset)

    # RMSE
    accuracy.rmse(predictions, verbose=True)
```

# ***CROSSFOLD VALIDATION***



*Crossfold validation* adalah teknik evaluasi model yang digunakan untuk mengukur kinerja model dengan membagi *dataset* menjadi beberapa *fold*, melatih model, dan  mengukur sejauh mana model SVD secara konsisten memberikan rekomendasi yang baik pada berbagai bagian *dataset*. *Crossfold validation* membantu memastikan bahwa model tidak hanya bekerja dengan baik pada satu bagian data tertentu, tetapi juga pada seluruh *dataset*, meningkatkan akurasi dan dapat mengidentifikasi apakah model rentan terhadap *overfitting*.







***K = 2***

```python
results2 = cross_validate(algo, data, measures=['MSE', 'RMSE'], cv=2, verbose=True)
```

# **TOP-N**

```python
# Fungsi untuk memberikan rekomendasi Top-N
def top_n(user_id, n=10):
    # List item
    all_items = list(trainset.all_items())                    # Semua item
    rated_items = [item for item, _ in trainset.ur[user_id]]  # Item yang sudah diberi rating
    unrated_items = list(set(all_items) - set(rated_items))   # Item yang belum diberi rating

    # Prediksi rating untuk item yang belum dinilai oleh pengguna
    predictions = [algo.predict(user_id, item) for item in unrated_items]
    predictions.sort(key=lambda x: x.est, reverse=True)

    # Mengambil 10 item teratas
    top_n = predictions[:n]
    author_name = reviews.loc[reviews['AuthorId'] == user_id]['AuthorName'].values[0]

    return top_n, author_name
```

```python
# Spesific User (AuthorId = 1634 || AuthorName = Bill Hilbrich)
user_id = 1634
top_n_recommendations, author_name = top_n(user_id, n=10)

print(f"Rekomendasi Top-N untuk Pengguna {user_id} ({author_name}):")
for i, prediction in enumerate(top_n_recommendations, 1):
    recipe_info = recipes.loc[recipes['RecipeId'] == prediction.iid]
    recipe_name = recipe_info['Name'].values[0] if not recipe_info.empty else "N/A"

    print(f"{i}. RecipeId: {prediction.iid}, Name: {recipe_name}")
    print(f"   Prediksi Rating: {prediction.est}")
```
